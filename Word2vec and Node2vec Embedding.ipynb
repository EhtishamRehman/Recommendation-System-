```python
import numpy as np
import pandas as pd
```

```python
#First we'll get the data and sort by time. The reason of sorting by time is due to the fact that we need the movies that each user saw in chronological order.
df_train = pd.read_csv("D:/University Stuff/Datasets/All Datasets/Movielens/u.data.csv", names=['user_id', 'item_id', 'ranking', 'time'], sep='\t')
df_train['time'] = pd.to_datetime(df_train['time'],unit='s')
df_train = df_train.sort_values(by='time')
```

```python
#And filter with ranking above of 3 points
df_train = df_train[df_train['ranking'] > 3]
```

```python
df_train.head()
```

```python
#Getting item features
names=['item_id', 'movie title', 'release date', 'video release date',
              'IMDb URL', 'unknown', 'Action', 'Adventure', 'Animation',
              'Children', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy',
              'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi',
              'Thriller', 'War', 'Western' ]
df_items = pd.read_csv("D:/University Stuff/Datasets/All Datasets/Movielens/u.item", names= names, sep='|', encoding = 'ISO-8859-1')
```

```python
df_items = df_items.filter(['item_id', 'movie title', 'Action', 'Adventure', 'Animation',
              'Children', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy',
              'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi',
              'Thriller', 'War', 'Western'])
```

```python
df_items.head()
```

```python
#Merge train with items to get their genre
df_train_2 = pd.merge(df_train, df_items, on='item_id')
```

```python
#Let's see if how the train dataframe was built. We want to see all the items watched by user_id 880
df_train_2[df_train_2['user_id'] == 880]
```

```python
#One film can be in several genres at once. We can see this adding a column 'Total' that is the sum of all the genres.

l = ['Action', 'Adventure', 'Animation', 
    'Children', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy',
    'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi',
    'Thriller', 'War', 'Western']
    
def f(row):
    sum = 0
    for i in l:
        sum = sum+row[i] 
    return sum
    
df_train_2['Total'] = df_train_2.apply(f, axis=1)
df_train_2.head()
```

```python
#In the following cell, we can see that the majority of the films have 2 features. 
#In the next steps we will see why this is important
df_items['Total'] = df_items.apply(f, axis=1)
df_items['Total'].describe()
```

```python
#Now we will create a dataset with all the users and all their watched films sorted by timestamp. Also we'll add the first genre. that is related to each movie. 
#I know that a movie has several genres but this is to keep it simple. Undoubtedly this could be coded better.

def get_features_movie(item_id):
    l = ['Action', 'Adventure', 'Animation', 
    'Children', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy',
    'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi',
    'Thriller', 'War', 'Western']
    
    features = []
    temp = df_items[df_items['item_id'] == item_id]
    for i in l:
        if temp.iloc[0][i]:
            features = features + [i]
    
    return features

def convert_to_list(item_id):
    return " ".join([str(x) for x in get_features_movie(item_id)[:1]])

train_watched = pd.DataFrame(columns=['user_id', 'watched'])

for index, user_id in enumerate(range(min(df_train_2['user_id']), max(df_train_2['user_id']))):
    d = df_train_2[df_train_2['user_id'] == user_id].filter(['item_id'])
    l = d['item_id'].tolist()
    l_to_string = " ".join([convert_to_list(x)+" "+str(x) for x in l])
    train_watched.loc[index] = [user_id, l_to_string]
```

```python
train_watched.head()
```

```python
#Therefore in the new dataset we can see the trace of watched films and its genre for each user. 
#It's time to use word2vec
from gensim.test.utils import common_texts
from gensim.models.word2vec import Word2Vec
```

```python
list_doc = []

for row in train_watched.to_dict(orient='record'):
    list_doc.append(str(row['watched']).strip().split(' '))
```

```python
model = Word2Vec(list_doc, window=5, min_count=1, workers=4)
#After train we have to see the results.
```

```python
def most_similar(item_id_or_genre):
    try:
        print("Similar of "+df_items[df_items['item_id'] == int(item_id_or_genre)].iloc[0]['movie title'])
    except:
        print("Similar of "+item_id_or_genre)
    return [(x, df_items[df_items['item_id'] == int(x[0])].iloc[0]['movie title']) for x in model.wv.most_similar(item_id_or_genre)]
```

```python
most_similar('Action')
```

```python
most_similar('Drama')
```

```python
#Let's see movies related with "Mirror Has Two Faces 1996"
most_similar('220')

#All the results are almost Romance/Drama, and also "Mirror Has Two Faces" is recommended. 
#Quite good! This could be used when you want a recommendation from a genre or an item. 
#But what about a recommendation for a user?. For example what films the user_id 880 saw?

```

```python
df_train_2[df_train_2['user_id'] == 880].filter(['item_id', 'movie title']+l)
#880 user saw these movies. 
```

```python
#Let's define a user as a vector of the average of movies that he saw.

def create_avg_user_vector(user_id):
    item_id_list = df_train_2[df_train_2['user_id'] == user_id]['item_id'].tolist()
    vector_item_id_list = [model.wv[str(x)] for x in item_id_list]
    return np.average(vector_item_id_list, axis=0)

def most_similar_by_vector(vector):
    return [(x, df_items[df_items['item_id'] == int(x[0])].iloc[0]['movie title']) for x in model.wv.similar_by_vector(vector)]

most_similar_by_vector(create_avg_user_vector(880))
```

```python
#Creating and analysing Graph
#Now, we will create a graph using movie titles and genres as a node using networkx. I use two functions rememeber ehtisham:-

#addToGraph(movie_name, graph): It adds an edge to the graph the edge contains the title and its genres as nodes
#createGraph(): This calls addToGraph for each movie title to create a complete graph
```

```python
# function that will create edges for given movie title and its genres
def addToGraph(movie_name,graph):
    genres=df[df['title']==movie_name]['listed_in'].values[0].rstrip().lower().split(', ')
    for genre in genres:
        graph.add_edge(movie_name.strip(),genre)
    return graph

#function that will create graph for all the movies name
def createGraph():
    graph = nx.Graph()
    for movie_name in df['title']:
        graph=addToGraph(movie_name,graph)
    return graph
```

```python
graph=createGraph()
```

```python
print(graph.degree()['Good People, September 8, 2017']) #should be 2 since two genres are assoicated with it
print(graph.degree()['Automata, September 8, 2017']) #shoukd be 3 since three genres are assoicated with it
```

```python
node2vec = Node2Vec(graph, dimensions=128, walk_length=60, num_walks=3)
```

```python
model = node2vec.fit(window=5, min_count=1)
```

```python
model.wv.get_vector('The Crow, January 1, 2019')
```

```python
#Using Node2Vec Embeddings
#I will use the embeddings generated to recommend similiar genres and movies. 
```

```python
def print_similiar(name):
    for node, _ in model.wv.most_similar(name):
        print(node)
```

```python
print_similiar('The Crow, January 1, 2019')
#As we can see most movies like "The Crow"
```
